```python
# Install Dependencies and Packages
!pip install stats-can openpyxl

import pandas as pd
import requests
import zipfile
from io import BytesIO
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
##############################
# Testing IMF API by fetching US and China GDP data
##############################

response = requests.get("https://www.imf.org/external/datamapper/api/v1/NGDP_RPCH/USA/CHN")

df_response_json = response.json()

print(df_response_json)
```

```python
############################
# Convert JSON into Data Frame
############################

#Extract nested data from the JSON, ensuring years are the index and countries are columns.


gdp_growth_data = df_response_json['values']['NGDP_RPCH']
df_gdp_growth = pd.DataFrame(gdp_growth_data)
df_gdp_growth.head()
```

```python
##############################
# Using IMF API to fetch GDP data for all countries
##############################

response2 = requests.get("https://www.imf.org/external/datamapper/api/v1/NGDPD")

df2_response_json = response2.json()

print(df2_response_json)

```

```python
#################################
# Convert the All-GDP JSON into a data frame
#################################

all_gdp_data = df2_response_json['values']['NGDPD']
df_all_gdp_data = pd.DataFrame(all_gdp_data)
df_all_gdp_data.head()
```

```python
################################
# Filter the new data frame to just 2025 for every country
################################

df_gdp_2025 = df_all_gdp_data[df_all_gdp_data.index == '2025']
df_gdp_2025.head()
```

```python
##################################
# Reformat as a long column for a table view
##################################

# Transpose the df_gdp_2025 DataFrame
df_gdp_2025_transposed = df_gdp_2025.T

# Rename the column containing the GDP values to '2025 GDP'
df_gdp_2025_transposed.rename(columns={'2025': '2025 GDP'}, inplace=True)

# Reset the index and rename the new column to 'Country'
df_gdp_2025_transposed.reset_index(inplace=True)
df_gdp_2025_transposed.rename(columns={'index': 'Country Code'}, inplace=True)

df_gdp_2025_transposed.head(200)
```

```python
#######################
# Replacing country codes with country names
#######################

## This API includes a dictionary of full country names for each code. Task is to take this and combine it with the existing table, replacing the codes.

countries = requests.get("https://www.imf.org/external/datamapper/api/v1/countries")
df_countries = pd.DataFrame(countries.json())

# Making codes a column instead of the index
df_countries.reset_index(inplace=True)
df_countries.rename(columns={'index': 'Country Code'}, inplace=True)

# Extract the country name from the dictionary in the 'countries' column
df_countries['countries'] = df_countries['countries'].apply(lambda x: x.get('label') if isinstance(x, dict) else x)

df_countries.head()
```

```python
######################
# Merging the country codes with GDP and country codes with full country names to get full country names and GDP in same data frame
######################

# Merging to create list of country with readable names instead of codes

pd.merge(df_gdp_2025_transposed, df_countries, on="Country Code", how="outer")
```

```python
#################################
# Pulling IMF codes for regions which for some reason are in a different API endpoint
#################################

response_regions = requests.get("https://www.imf.org/external/datamapper/api/v1/regions")
df_regions = pd.DataFrame(response_regions.json())

# Making codes a column instead of the index
df_regions.reset_index(inplace=True)
df_regions.rename(columns={'index': 'Country Code', 'regions': 'countries'}, inplace=True)

# Extract the country name from the dictionary in the 'regions' column
df_regions['countries'] = df_regions['countries'].apply(lambda x: x.get('label') if isinstance(x, dict) else x)

df_regions.head(30)
```

```python
#######################
# Ranking from highest to lowest by country GDP
#######################

df_GDP = pd.merge(df_gdp_2025_transposed, df_countries, on="Country Code", how="outer")
df_GDP_all = pd.merge(df_GDP, df_regions, on="Country Code", how="outer")

# Display every value for quick visibiilty
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Sort highest to lowest and display
ranked_gdp = df_GDP_all.sort_values("2025 GDP", ascending=False)

print(ranked_gdp)

pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

```python
# Drop the 'api' column
ranked_gdp_cleaned = ranked_gdp.drop(columns=['api'])

# Reset the index and drop the old one
ranked_gdp_cleaned = ranked_gdp_cleaned.reset_index(drop=True)

# Display the cleaned DataFrame
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
print(ranked_gdp_cleaned)
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```
